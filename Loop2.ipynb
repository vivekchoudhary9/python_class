```python
import math
s = int(input("Enter starting point : "))
e = int(input("Enter Ending Point : "))
total_prime = 0
while s <= e :
    num = s
    end = int(math.sqrt(num)) + 1
    start = 2
    while start <= end :
        if num % start == 0 :
            break
        start = start + 1
    else :
        total_prime += 1
        print(num,end='\t')
    s = s + 1
print("\nTotal Prime : ",total_prime)
```

```python
import math
s = int(input("Enter starting point : "))
e = int(input("Enter Ending Point : "))
total_prime = 0
for k in range(s,e+1):
    num = k
    end = int(math.sqrt(num)) + 1
    start = 2
    for t in range(start,end+1) :
        if num % t == 0 :
            break
        
    else :
        total_prime += 1
        print(num,end='\t')
print("\nTotal Prime : ",total_prime)
```
